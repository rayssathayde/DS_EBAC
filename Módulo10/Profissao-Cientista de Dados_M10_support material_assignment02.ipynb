````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
previsao_de_renda = pd.read_csv('previsao_de_renda.csv')
previsao_de_renda
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
plt.rc('figure', figsize=(10,8))
fig, axes = plt.subplots(2,2)

sns.pointplot(ax=axes[0,0],
              x='posse_de_veiculo', y='renda', 
              data= previsao_de_renda, dodge= True, 
              errorbar=('ci', 90))

sns.pointplot(ax=axes[0,1],
              x='posse_de_imovel', y='renda', 
              data= previsao_de_renda, dodge= True, 
              errorbar=('ci', 90))

sns.pointplot(ax=axes[1,0],
              x='posse_de_veiculo', y='renda', 
              data= previsao_de_renda, dodge= True, 
              hue='posse_de_imovel', errorbar=('ci', 90))

sns.pointplot(ax=axes[1,1],
              x='posse_de_imovel', y='renda', 
              data= previsao_de_renda, dodge= True, 
              hue='posse_de_veiculo', errorbar=('ci', 90))
````

As duas variáveis qualitativas (``posse de veiculo`` e ``posse de imovel``) são boas preditoras de renda e parecem discriminar renda de forma estatisticamente significante.
Nos gráficos da primeira linha podemos observar que, na comparação das médias, as pessoas que possuem imóvel e as que possuem veículo apresentam maior renda em comparação às que não possuem.
Observando os gráficos da segunda linha podemos perceber que as pessoas que possuem veículo, mas não possuem imóvel, apresentam maior renda do que as que possuem imóvel mas não possuem veículo. Dessa forma, podemos observar que a variável ``posse de veiculo`` tem maior valor preditivo de renda. Além disso, as pessoas que possuem veículo apresentam renda parecida, independente de possuírem ou não imóvel.
