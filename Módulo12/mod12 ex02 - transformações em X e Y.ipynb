# Regressão 01 - tarefa 02 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf
````

````python
# carregar base "tips" do seaborn

tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
# modelo de regressão

reg1 = smf.ols('tip_pct ~ net_bill', data=tips).fit()
print(reg1.summary())

tips['res'] = reg1.resid

sns.scatterplot(x = 'net_bill', y = 'res', data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
# gráfico de dispersão 

sns.regplot(y = 'tip_pct', x = 'net_bill', data = tips[tips['tip_pct']<.5])
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x = 'net_bill', y = 'res', data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')

plt.show()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg2 = smf.ols('tip_pct ~ net_bill + np.power(net_bill, 2)', data=tips).fit()
print(reg2.summary())

tips['res'] = reg.resid

sns.scatterplot(x = 'net_bill', y = 'res', data = tips, alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg2.fittedvalues, 'r,--')

plt.show()
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg3 = smf.ols('np.log(tip_pct) ~ net_bill', data=tips).fit()
print(reg3.summary())

tips['res'] = reg3.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
# não podemos comparar os r-quadrado, precisamos passar o tip_pct para escala anterior

tips['pred_tip_pct'] = np.exp(reg3.fittedvalues)

print('R-squared:', 
      tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
     )
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')

plt.show()
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg4 = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data=tips).fit()
print(reg4.summary())

tips['res'] = reg4.resid

sns.scatterplot(x='net_bill', y='res', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
# não podemos comparar os r-quadrado, precisamos passar o tip_pct para escala anterior

tips['pred_tip_pct'] = np.exp(reg4.fittedvalues)

print('R-squared:',
      tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
     )
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')

plt.show()
````
