````python
#1) Seu código aqui

````

````python
def consulta (UF_interessado):
    r = requests.get(UF_interessado)
    print(r)
    data_json = r.json()
    df = pd.DataFrame(data_json['items'])
    return df 
    
````

````python
RJ_municipio = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'
````

````python
consulta (RJ_municipio)
````
