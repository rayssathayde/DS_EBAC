# Probabilidade - Tarefa

```python
import math
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
media = 15  # média
desvio_padrao = 0.1  # desvio padrão

# probabilidade de que um pacote tenha mais que 15.2 kg
prob = 1 - stats.norm.cdf(15.2, media, desvio_padrao)

print("A probabilidade de que um pacote selecionado aleatoriamente tenha mais que 15.2 kg é de {:.2%}".format(prob))
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# percentil de ordem 95% para o peso dos pacotes
perc_95 = stats.norm.ppf(0.95, media, desvio_padrao)

print("O percentil de ordem 95% para o peso dos pacotes é de {:.2f} kg".format(perc_95))
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y = np.linspace(-6, 6, 200)

# probabilidade Y1

media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

# probabilidade de Y1 ser menor que cada valor em Y (distribuição normal, média 0 e variância 1)
Y1 = stats.norm.cdf(x=Y, loc=media, scale=desvio_padrao)

# probabilidades para Y2 (distribuição t-student com 5 graus de liberdade)
Y2 = stats.t.cdf(Y, df=5)

#colocar todos os valores em um df
data = {"Y": Y, "Probabilidade_Y1": Y1, "Probabilidade_Y2": Y2}
df = pd.DataFrame(data)
df = df.round(4)

df
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Criação da figura e do eixo
fig, ax = plt.subplots(figsize=(30, 9))

# Plotagem dos dados de Y1 e Y2
ax.plot(Y, Y1, label='Y1 - Normal')
ax.plot(Y, Y2, label='Y2 - t(5)')

# Configuração dos marcadores no eixo x e y
ax.xaxis.set_major_locator(plt.MaxNLocator(14))
ax.yaxis.set_major_locator(plt.MaxNLocator(11))

# Formatação do eixo y em formato percentual
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))

# Adição de rótulo para os eixos 
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidade Acumulada')

# Adição da legenda
ax.legend()

# Exibição do gráfico
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P = np.linspace(0, 1, 20)

# probabilidade Y1

media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

# percentil de Y1 para cada valor em P (distribuição normal, média 0 e variância 1)
Y1_per = stats.norm.ppf(P, media, desvio_padrao)

# percentil de Y2 para cada valor em P (distribuição t-student com 5 graus de liberdade)
Y2_per = stats.t.ppf(P, df=5)

#colocar todos os valores em um df
data1 = {"P": P, "Percentil_Y1": Y1_per, "Percentil_Y2": Y2_per}
df1 = pd.DataFrame(data1)
df1 = df1.round(4)

df1
```

6. Faça um gráfico com os dados do item anterior.

```python
# Criação da figura e do eixo
fig, ax = plt.subplots(figsize=(30, 9))

# Plotagem dos dados de Y1 e Y2 em função dos percentis P
ax.plot(P, Y1_per, label='Y1 - Normal')
ax.plot(P, Y2_per, label='Y2 - t(5)')

# Configuração dos marcadores no eixo x e y
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.yaxis.set_major_locator(plt.MaxNLocator(22))

# Adição de rótulo para os eixos 
ax.set_xlabel('Percentis (P)')
ax.set_ylabel('Valores de Y1 e Y2')

# Adição da legenda
ax.legend()

# Exibição do gráfico
plt.show()
```
