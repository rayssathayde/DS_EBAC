#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
data = sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
data
```

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=data)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=data)
sns.countplot(ax = axes[1, 0], x='SEXO', data=data)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=data)
sns.countplot(ax = axes[2, 0], x='ESTCIVMAE', data=data)

plt.subplots_adjust(wspace=0.4, hspace=0.4)
axes[1,1].tick_params(axis="x", rotation=90)

plt.axis('off')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(12, 4))
fig, axes = plt.subplots(1,2)

sns.countplot(ax = axes[0], x='APGAR5',data=sinasc)
sns.barplot(ax = axes[1], x='APGAR5', y='CONSULTAS', data=sinasc)

```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2,2)

sns.countplot(ax = axes[0, 0], x='CONSULTAS', data=sinasc)
sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
axes[1,1].tick_params(axis="x", rotation=90)
```
