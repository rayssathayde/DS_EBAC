````python
import pandas as pd

import plotly.express as px 
from plotly import graph_objects
from plotly.graph_objs import Layout
````

````python
df = pd.read_csv('ativos.csv', header=[0,1], index_col=0)
df.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# média móvel dos ativos ao longo do tempo
pd.set_option('display.max_rows', None)
media_movel = df['Adj Close'].rolling(10).mean()
media_movel
````

Como tinham alguns valores nulos ao longo do df, quando coloquei window=90 apareceu praticamente tudo NaN.<br> Então coloquei um intervalo menor.

````python
# desvio padrão dos ativos ao longo do tempo
pd.set_option('display.max_rows', 10)
desvio_padrao = df['Adj Close'].rolling(10).std()
desvio_padrao
````

````python
media_movel_vert = media_movel.stack().reset_index()
media_movel_vert.columns=['data', 'ativo', 'valor']
media_movel_vert
````

````python
fig = px.line(media_movel_vert, 
              x='data', 
              y='valor',
              color='ativo', 
              template='simple_white')
fig.show()
````

````python
desvio_padrao_vert = desvio_padrao.stack().reset_index()
desvio_padrao_vert.columns=['data', 'ativo', 'valor']
desvio_padrao_vert
````

````python
fig = px.line(desvio_padrao_vert, 
              x='data', 
              y='valor',
              color='ativo', 
              template='simple_white')
fig.show()
````

No gráfico do exercício anterior os ativos da Nestlé (NSRGF) apresentavam a maior volatilidade. Nesses gráficos podemos perceber que os ativos da Garmin também apresentam certa volatilidade, e a Natura parece mais estável nos 3 gráficos. 

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['NTCO3.SA']*len(lista_indicadores)
````

````python
colunas = list(zip(lista_indicadores, ativo))
colunas
````

````python
ntco = df.loc[:,colunas]
ntco.columns = lista_indicadores
ntco = ntco.tail(60)
ntco
````

````python
layout = Layout(paper_bgcolor='rgba(0,0,0,0)', 
                plot_bgcolor='white', 
                height=600)

graph = {
    'x': ntco.index, 
    'open': ntco.Open,
    'close': ntco.Close,
    'high': ntco.High,
    'low': ntco.Low,
    'type': 'candlestick',
    'name': 'NTCO3.SA',
    'showlegend': True
}

grafico = graph_objects.Figure(data=[graph], 
                               layout_title='Índice Natura', 
                               layout=layout)

grafico.update_xaxes(showgrid=True, 
                     gridwidth=1, 
                     gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, 
                     gridwidth=1, 
                     gridcolor='LightGrey')

grafico
````
