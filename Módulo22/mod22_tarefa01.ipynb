#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
yf.head(2)
```

```python
yf.columns
```

```python
#amplitude de variação do valor do ativo dentro do dia

yf['Range', 'GRMN'] = yf['High', 'GRMN'] - yf['Low', 'GRMN']
yf['Range', 'NSRGF'] = yf['High', 'NSRGF'] - yf['Low', 'NSRGF']
yf['Range', 'NTCO3.SA'] = yf['High', 'NTCO3.SA'] - yf['Low', 'NTCO3.SA']
yf
```

```python
yrange = yf['Range']
yrange_vert = yrange.stack().reset_index()
yrange_vert.columns=['data', 'ativo', 'valor']
yrange_vert
```

```python
fig = px.line(yrange_vert, 
              x='data', 
              y='valor',
              color='ativo', 
              template='simple_white')
fig.show()
```

Entre as empresas Natura (NTCO3.SA), Nestlé (NSRGF) e Garmin (GRMN), a Nestlé apresenta maior volatividade.
