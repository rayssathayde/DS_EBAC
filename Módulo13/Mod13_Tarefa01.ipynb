```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy


%matplotlib inline
```

```python
renda = pd.read_csv('previsao_de_renda.csv')
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.   

```python
renda.columns
```

```python
# verificar a quantidade de registros dos valores em cada variável
# utilizar o registro mais frequente como casela de referência 

colunas = renda.columns

for col in colunas:
    print(renda[col].value_counts())
```

```python
y, X = patsy.dmatrices('''np.log(renda) ~ sexo 
                                        + posse_de_veiculo
                                        + posse_de_imovel
                                        + C(qtd_filhos, Treatment(0)) 
                                        + C(tipo_renda, Treatment('Assalariado')) 
                                        + C(educacao, Treatment('Secundário'))
                                        + C(estado_civil, Treatment('Casado'))
                                        + C(tipo_residencia, Treatment('Casa'))
                                        + idade
                                        + tempo_emprego
                                        + C(qt_pessoas_residencia, Treatment(2))''', renda)
```

```python
reg = sm.OLS(y, X).fit()
reg.summary()
```

As variáveis ``qtd_filhos``, ``tipo_renda``, ``educacao``, ``estado_civil`` e ``qt_pessoas_residencia`` apresentam categorias que são relevantes e outras que não são. 
A variável ``tipo_residencia`` não apresenta nenhuma categoria relevante. 

2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.

```python
# remover tipo_residencia, a variável menos significante 

y, X = patsy.dmatrices('''np.log(renda) ~ sexo 
                                        + posse_de_veiculo
                                        + posse_de_imovel
                                        + C(qtd_filhos, Treatment(0)) 
                                        + C(tipo_renda, Treatment('Assalariado')) 
                                        + C(educacao, Treatment('Secundário'))
                                        + C(estado_civil, Treatment('Casado'))
                                        + idade
                                        + tempo_emprego
                                        + C(qt_pessoas_residencia, Treatment(2))''', renda)
```

```python
reg = sm.OLS(y, X).fit()
reg.summary()
```

R-quadrado permaneceu o mesmo. AIC diminuiu muito pouco, mas diminuiu. R-quadrado ajustado permaneceu o mesmo. Aparentemente o modelo não sofreu nenhuma mudança perceptível.
Os valores de p estão muito parecidos também, sem mudanças significativas. 

3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 

```python
# retirei variáveis em que a maioria das categorias não são significativas - qtd_filhos, educacao e qt_pessoas_residencia
# mantive algumas categorias - significativas - de outras variáveis - tipo_renda 

y, X = patsy.dmatrices('''np.log(renda) ~ sexo 
                                + posse_de_veiculo 
                                + posse_de_imovel
                                + tipo_renda == 'Empresário'
                                + idade 
                                + tempo_emprego''', renda)
```
